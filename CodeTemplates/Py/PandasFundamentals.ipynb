```python
#Challenge 4

genderList = []
sList = df['sex']

for s in sList:
    if not s in genderList:
        genderList.append(s)

print(str(len(genderList)), "genders in this dataset.")
```

```python
#Challenge 4

print(str(df['age'].sum()/len(df['age'])), "is the average age of all pirates.")

fAges = []
i=0
for s in df['sex']:
    if s == "female":
        fAges.append(float(df['age'][i]))
    i+=1
        
total = 0
for age in fAges:
    total+=age

print(str(total/len(fAges)), "is the average age of all female pirates.")
print(str(df['age'][df['sex']=="female"].mean()))
```

```python
#Challenge 4

count = 0
for w in df['weight']:
    if w < 45.3592:
        count+=1
        
print(str(count/len(df['weight'])*100)+"% of pirates were less than 100 pounds.")
print(len(df['weight'][df['weight']<45.3592])/len(df['weight'])*100)
```
