```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://www.scrapethissite.com/pages/simple/"

response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, 'html.parser')

print(soup)
```

```python
countries = soup.find_all('h3', class_='country-name')

country_names = [country.get_text().strip() for country in countries]
capitals = [item.get_text() for item in soup.find_all('span', class_='country-capital')]
populations = [item.get_text() for item in soup.find_all('span', class_='country-population')]
areas = [item.get_text() for item in soup.find_all('span', class_='country-area')]

print(country_names)
print("Capitals:", capitals)
print("Populations:", populations)
print("Areas:", areas)
```

```python
df = {
    "Country": country_names,
    "Capital": capitals,
    "Population": populations,
    "Area": areas
}
x = pd.DataFrame(df)
x
```
